```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
```

```python
df = pd.read_csv("anemia.csv")
df.head(3)
```

```python
y=df["Diagnosis"]
x=df.drop(columns=["Diagnosis"])
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
le.inverse_transform([4])
```

```python
y = to_categorical(y)
```

```python
y[0]
```

```python
xC = len(x.columns)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.75,random_state=43)
```

```python
model = Sequential()

#input layer
model.add(Dense(256,activation="relu",input_dim =xC))

#hidden layer
model.add(Dense(128,activation="relu"))
model.add(Dense(64,activation="relu"))
model.add(Dense(64,activation="relu"))

#output layer

model.add(Dense(9,activation="softmax"))
```

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit(x_train,y_train,epochs=128,batch_size=32,validation_split=0.24)
```

```python
tahmin = np.array([[9.4,34,58,2.8,4.3,3.3,8.8,32,78,24,26,155,15,0.17]])
sonuc = model.predict(tahmin)
```
